```python
# Exercise 2.4: Time Dilation
# Notes on Lorentz Transformations (special relativity)
    # https://en.wikipedia.org/wiki/List_of_relativistic_equations
# Lorentz Factor (gamma): gamma = 1/sqrt(1-beta^2)
    # beta = v/c, v = relative velocity between inertial frames
# Time dilation: t' = gamma*t (ie "proper time" * the lorentz factor is the dilated time)
# Length contraction: l' = l/gamma (similar story, proper length / lorentz factor is new length)

x = float(input("Enter the distance between Earth and the destination (ly): "))
v = float(input("Enter the ship's relativistic speed as a fraction of the speed of light: "))
C = 2.998e8
L = 1/((1-(v)**2)**(1/2)) # Lorentz factor calculation, v*c / c = beta
pT = x/v # proper time, experienced from the relative frame
rT = pT*L # relative time, experienced from the rest frame

print("A passenger will experience the trip as", pT, "years. An observer on Earth will see the trip take",rT,"years.")
```

```python
# Exercise 2.5: Quantum Potential Step
# When faced with a potential step of height V, a particle with energy E such that E > V will either:
    # a) make the jump, (transmission, T), taking on energy of E - V and a wavevector k2 = sqrt(2m(E-V))/hbar
    # b) reflect (R), retaining it's energy but moving in the opposite direction
# The particle initially has wavevector k1 = sqrt(2mE)/hbar
# Probability of T = 4k1k2/(k1+k2)^2
# Probability of R = ((k1-k2)/(k1+k2))^2
# Equations drawn from SchrÃ¶dinger's equation

m = float(input("Enter the particle's mass (kg): "))
E = float(input("Enter the particle's energy (eV): "))
V = float(input("Enter the height of the potential step (eV): "))
hbar = 1.054571817e-34
k1 = (2*m*E)**(1/2)/hbar
k2 = (2*m*(E-V))**(1/2)/hbar
T = (4*k1*k2)/(k1+k2)**2
R = ((k1-k2)/(k1+k2))**2
print("The probability of transmission is ", T,". The probability of reflection is ", R, sep="")
```

```python
from math import pi
# Exercise 2.6: Planetary Orbits
# by Kepler's laws, l2v2 (aphelion) = l1v1 (perihelion)
# aphelion is when the distance is at it's greatest, so l2 > l1. Therefore, v2 < v1
# semimajor axis: a = 1/2*(l1+l2)
# semiminor axis: b = sqrt(l1l2)
# orbital period: T = 2*pi*a*b/(l1v1)
# orbital eccentricity (stretched out): e = (l2-l1)/(l2+l1)

l1 = float(input("Enter the distance to the Sun at perihelion (m): "))
v1 = float(input("Enter the velocity of the body relative to the Sun at perihelion (m/s): "))
G = 6.6738e-11
M = 1.9891e30
sty = 3.154e7
v2 = (2*G*M/(v1*l1)-((2*G*M/(v1*l1))**2-4*(2*G*M/l1-v1**2))**(1/2))/2 # by conservation of kinetic energy between perihelion and aphelion
l2 = l1*v1/v2
a = 1/2*(l1+l2)
b = (l1*l2)**(1/2)
T = 2*pi*a*b/(l1*v1)
e = (l2-l1)/(l2+l1)
print("The velocity at aphelion is {} m/s. The distance from the Sun at aphelion is {} m. The orbital period is therefore {} years, and the eccentricity is {}.".format(v2, l2, T/sty, e))
```

```python
# Exercise 2.9: The Madelung Constant
# total electric potential felt by atom in a solid. evaluate for sodium chloride in a cube w/ side length L
# i,j,k coordinates, distance from atom at i,j,k from origin is a*sqrt(i^2+j^2+k^2) for spacing of atoms a
# potential V(i,j,k) = +- e/(4*pi*epsilon0*a*sqrt(i^2+j^2+k^2)
    # epsilon0 = permittivy of the vacuum
    # +-, positive if i+j+k is even, negative if it's odd
# so the total potential is the sum of V(i,j,k) for all (i,j,k) in L^3
# = e/(4*pi*epsilon0*a)*M, where M is the Madelung Constant
# solving gives M = the sum of all +-1/(i^2+j^2+k^2)
from numpy import abs
M = 0
L = int(input("Please input the side length of the cubic volume in atoms: "))
temp = 0
for i in range(-L, L):
    for j in range(-L, L):
        for k in range(-L, L):
            if i==j==k==0: continue
            temp = i**2+j**2+k**2
            if temp % 2 == 0: M += 1/(temp**(1/2))
            else: M -= 1/(temp**(1/2))
print("The Madelung constant of NaCl is approximately", abs(M))
```

```python
# Exercise 2.13: Recursion

# a) Catalan Numbers
# C(n) = 1 if n = 0, (4n-2)/n+1 * C(n-1) if n > 0
def catalan(n):
    if n == 0: return 1
    else:
        return (4*n-2)/(n+1) * catalan(n-1)
print(catalan(100))

# b) Euclid's Algorithm (gcd)
# gcd(m, n) = m if n = 0, gcd(n, m mod n) if n > 0
def gcd(m, n):
    if n == 0: return m
    else:
        return gcd(n, m % n)
print(gcd(108, 192))
```

```python
# Exercise 3.6: Deterministic Chaos and the Feigenbaum plot
# x' = rx(1-x) for a constant r and an initial x. Compute x', the next data point uses that value as x
# Iterative map. Will result in 1 of 3 outcomes
    # 1. settles to a fixed point
    # 2. settles into a periodic pattern
    # 3. deterministic chaos
from pylab import scatter, show, xlabel, ylabel
from numpy import arange, repeat
def iMap(r, x):
    return r*x*(1-x)

r = arange(1,4,0.01)
for i in r:
    x = [1/2]
    for j in range(100):
        x.append(iMap(i, x[j]))
    ri = repeat(i, len(x)-1)
    scatter(ri, x[1:])

xlabel("r")
ylabel("x")
show()

# a) we see fixed points as stable values of x for a given r. limit cycles will be where we see values of x for each r
# surrounding a void, oscillating from one side to the other. Chaos is a bunch of points filling the plot
# b) we can identify the edge of chaos as approximately r=3.5
```

```python
# Exercise 3.7: Mandelbrot Set
# z' = z^2 + c, for z a complex number and c a complex constant.
# A point is a member of the Mandelbrot Set if |z| <= 2
# c = x + iy
from numpy import absolute, arange
from pylab import imshow, hot, show

def mandelbrot(z, c, val):
    while absolute(z) <= 2 and val < 100:
        return mandelbrot(z**2+c, c, val+1)
    return val

N = 1000
points = []
xi = arange(-2, 3, 1/(N/5))
yi = arange(-2, 3, 1/(N/5))

for x in range(N):
    temp = [0]*N
    for y in range(N):
        val = mandelbrot(0, complex(xi[x], yi[y]), 0)
        temp[y] = val
    points.append(temp)
imshow(points, origin="lower")
hot()
show()
```

```python
# Exercise 3.8: Least-squares fitting and the photoelectric effect
# Method of least-squares: a manner of finding a line of best fit (solving for m and c in y = mx + c)
# Ex = 1/N * sum(all x), Ey = 1/N * sum(all y), Exx = 1/N * sum(all x^2), Exy = 1/N * sum(all x * all y)
# m = (Exy - ExEy)/(Exx-(Ex)^2), c = (Exx*Ey-Ex*Exy)/(Exx - (Ex)^2)
# use millikan.txt, giving frequencies and voltages from photoelectric measurements, to approximate planck's constant
from pylab import xlabel, ylabel, plot, show
from numpy import loadtxt
def sums(x, y, N):
    Ex = 1/N * sum(x)
    Ey = 1/N * sum(y)
    Exx = 1/N * sum(x**2)
    Exy = 1/N * sum(x*y)
    return Ex, Ey, Exx, Exy
data = loadtxt("files/millikan.txt", float)
x = data[:, 0]
y = data[:, 1]

Ex, Ey, Exx, Exy = sums(x, y, len(x))
m = (Exy-Ex*Ey)/(Exx - Ex**2)
c = (Exx*Ey-Ex*Exy)/(Exx-Ex**2)
print("Slope:",m, "\tIntercept:",c)

line = []
for i in range(len(x)):
    line.append(m*x[i]+c)

# V = h/e * v - phi
# we see that h/e is equivalent to the slope of the graph. We can therefore calculate Planck's constant
e = 1.602e-19
h = m*e
print("Approx Planck's constant:", h)

plot(x, y, "ko")
plot(x, line)
xlabel("Frequency (Hz)")
ylabel("Voltage (V)")
show()
```
